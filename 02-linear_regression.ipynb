### Notes:
Regression is a technique to predict new data based on previous data.
The simplest way to do this is to assume that the correlation between known and future data is linear. That is why it is called Linear Regression
We can think of our equation as  Y=g(x) . Y is our output and x is the input.  g  is a linear function that acts on x. This is how we predict the data.
Finding  g  is the same as finding the line of best fit which we can do using code.

### Finding the Function:
- The math that is used to find the line of best fit is complex and can take a long time to solve by hand when there is a lot of data points.
- To save time, we can use Python modules to do the work for us.

```python
import numpy as np
import matplotlib.pyplot as plt
# Hours studied (independent variable)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
# Exam scores (dependent variable)
Y = np.array([60, 62, 63, 65, 70])

### Visualize the data
plt.scatter(X, Y, color='blue')
plt.title('Hours Studied vs. Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()
```

We can notice obvious linear relationship, so linear regression is a good option!
Let's build a linear model for it!

```python
### Step 1: import linear model

from sklearn.linear_model import LinearRegression

# regression model
reg = LinearRegression()



# Step 2: fit (model training)
reg.fit(X, Y)

# Step 3: predict
F = reg.predict([[3.5]])

# Create a scatter plot of the data points
plt.scatter([x[0] for x in X], Y, color='blue', label='training data')

# Create the regression line
X_range = [[min([x[0] for x in X])], [max([x[0] for x in X])]]
Y_range = reg.predict(X_range)
plt.plot([x[0] for x in X_range], Y_range, color='red', label='Regression line')

# Plot the predicted value
plt.scatter([3.5], F, color='green', marker='x', s=100, label='Predicted Score')

# Add labels and title
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Hours Studied vs. Exam Score')
plt.legend()

# Show plot
plt.show()
```
