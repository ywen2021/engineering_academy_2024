## Variables

Like we would do in mathematics when we define variables in equations such as $x=3$, we can do the same in all programming languages. Python has one of the simplest syntax for this, i.e. exactly as we would do it naturally. Let's define a variable in the next cell:

````python
a = 3
a
````

Some other types of data, for example we can define a **text** variable by using quotes ```' '``` around a given text:

````python
my_text = 'This is my text'
my_text
````

Others can contain multiple elements like lists:

````python
my_list = [3, 8, 5, 9]
my_list
````

## Functions
If we want to go beyond simple arithmetic we need more **complex functions** that can operate on variables. Imagine for example that we need a function $f(x, a, b) = a * x + b$. For this we can use and **define functions**. Here's how we can define a  function:

````python
def my_fun(x, a, b):
    out = a * x + b
    return(out)
````

We can now use this function as if we were doing mathematics: we pick a a value for the three parameters e.g. $f(3, 2, 5)$

````python
my_fun(3, 2, 5)
````

Another example, define a function for greetings

````python
def greet(name):
    return(f"Hello, {name}!")
````

````python
# Calling this function
print(greet("Alice"))
````

## Packages

In the examples above, we defined a function ourselves but there are some existing functions: **external packages**. These packages are collections of functions used in a specific domain that are made available to everyone via specialized online repositories. For example we will be using in this course a package called [scikit-image](https://scikit-image.org/) that implements a large number of functions for image processing.

### Importing functions

The answer is that we have to **import** the functions we want to use in a *given notebook* from a package to be able to use them. First the package needs to be **installed**.  We can install packages from there using the following command either in a **terminal or directly in the notebook**. For example for [scikit-image](https://pypi.org/project/scikit-image/):


````python
!pip install scikit-image
````

Note that, Google colab comes with many pre-installed libraries. So for most packages we don't need to install. Hooray!!!

Once installed we can **import** the packakge in a notebook in the following way (note that the name of the package is scikit-image, but in code we use an abbreviated name ```skimage```):

````python
import skimage
````

The import is valid for the **entire notebook**, we don't need that line in each cell.

Now that we have imported the package we can access all function that we define in it using a *dot notation* ```skimage.myfun```. Most packages are organized into submodules and in that case to access functions of a submodule we use ```skimage.my_submodule.myfun```.

Note that there are multiple ways to import packages. For example we could give another name to the package, using the ```as``` statement:

````python
import skimage as sk
````

We can also import only a certain submodule using:

````python
from skimage import filters
````

Or import a **single** function like this:

````python
from skimage.filters import median
````

Let's check how we read an image:

````python
image_address = 'https://github.com/guiwitz/PyImageCourse_beginner/raw/master/images/19838_1252_F8_1.tif'
````

Now we can call our function:

````python
skimage.io.imread(image_address)
````

We see here an output which is what is returned by our function. It is as expected a list of numbers, and not all numbers are shown because the list is too long. We see that we also have ```[]``` to specify rows, columns etc. The main difference compared to our list of lists that we defined previously is the ```array``` indication at the very beginning of the list of numbers. This ```array``` indication tells us that we are dealing with a ```Numpy``` array, this alternative type of list of lists that will allow us to do efficient computations.

## Plotting

For plotting we will use yet another **external library** called Matplotlib. That library is extensively used in the Python world and offers extensive choices of plots. We will mainly use one **function** from the library to display images: ```imshow```. Again, to access that function, we first need to import the package. Here we need a specific submodule:

````python
import matplotlib.pyplot as plt
````

Now we can use the ```plt.imshow()``` function. There are many options for plot, but we can use that function already by just passing an ```array``` as an input. First we need to assign the imported array to a variable:

````python
import skimage.io

image = skimage.io.imread(image_address)
````

````python
plt.imshow(image);
````
